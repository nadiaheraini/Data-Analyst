```python
#Import library
import pandas as pd
import numpy as np
```

```python
superstore = pd.read_csv("https://raw.githubusercontent.com/densaiko/data_science_learning/main/dataset/Sample-Superstore%20-%20Orders.csv")
superstore.head()
```

```python
sorting_data = superstore.sort_values(by=['Sales'], ascending=True)
sorting_data.head(10)
```

```python
sorting_data = superstore.sort_values(by=['Sales'], ascending=False)
sorting_data.head(10)
```

```python
sorting_data = superstore.sort_values(by=['Ship Mode', 'Segment', 'Sales'], ascending=[False, False, False])
sorting_data.head(10)
```

```python
melting_data = pd.melt(superstore, id_vars=['Category'], value_vars=['Sales'])
melting_data
```
